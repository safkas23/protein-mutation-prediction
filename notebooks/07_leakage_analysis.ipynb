```python
import pandas as pd
from difflib import SequenceMatcher
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")

train_seqs = set(train_df["seq"])
test_seqs = set(test_df["seq"])
```

```python
exact_overlap = len(
    train_seqs.intersection(test_seqs)
)

print("Exact overlap:", exact_overlap)
```

```python
def seq_similarity(a,b):
    return SequenceMatcher(
        None,
        a,
        b
    ).ratio()

similar_pairs=0
for test_seq in list(test_seqs)[:100]:

    for train_seq in list(train_seqs)[:500]:

        if seq_similarity(
            test_seq,
            train_seq
        ) > .90:

            similar_pairs += 1
            break

print("Near duplicates:", similar_pairs)
```

```python
results = pd.DataFrame([
{
"Exact_Overlap":exact_overlap,
"Near_Duplicates":similar_pairs
}
])

results.to_csv(
"../results/leakage_analysis.csv",
index=False
)
```
